```python
import matplotlib.pyplot as plt
%matplotlib inline

import cv2
```

## 使用matplot

```python
img = cv2.imread('.\\img\\rabbit1.jpg', cv2.IMREAD_COLOR)
print(img.shape)

plt.imshow(img)
```

```python
img = cv2.imread('.\\img\\rabbit1.jpg', cv2.IMREAD_GRAYSCALE)
print(img.shape)

plt.imshow(img)
```

```python
img = cv2.resize(img, (0,0), fx=3, fy=1)
print(img.shape)

plt.imshow(img)
```

```python
img = cv2.resize(img, (500, 200))
print(img.shape)

plt.imshow(img)
```

## 非matplot方式

```python
def imshow(name, img):
    cv2.imshow(name, img)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
```

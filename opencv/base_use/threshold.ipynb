## 阈值

cv2.threshold: src, thresh:阈值, maxval:最大值, type:二值化操作类型
+ cv2.THRESH_TRUNC If value large than thresh, value:=thresh
+ cv2.THRESH_BINARY If value large than thresh, value:=maxval or value:=0
+ cv2.THRESH_BINARY_INV ~cv2.THRESH_BINARY
+ cv2.THRESH_TOZERO If value less than thresh, value:= 0
+ cv2.THRESH_TOZERO_INV ~cv2.THRESH_TOZERO
+ cv2.THRESH_MASK
+ cv2.THRESH_OTSU
+ cv2.THRESH_TRIANGLE 三角形最大直线距离

```python
import cv2
```

```python
def imshow(name, img):
    cv2.imshow(name, img)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
```

```python
rabbit = cv2.imread('.\\img\\rabbit1.jpg', cv2.IMREAD_COLOR)
cat = cv2.imread('.\\img\\cat1.jpg', cv2.IMREAD_COLOR)
print(rabbit.shape, cat.shape)
```

```python
# img = cv2.addWeighted(rabbit, 0.4, cat, 0.6, 0)
img = cv2.cvtColor(rabbit, cv2.COLOR_BGR2GRAY)
ret1, trunc = cv2.threshold(img, 127, 255, type=cv2.THRESH_TRUNC)
ret2, binary = cv2.threshold(img, 127, 255, type=cv2.THRESH_BINARY)
ret3, binary_inv = cv2.threshold(img, 127, 255, type=cv2.THRESH_BINARY_INV)
ret4, zero = cv2.threshold(img, 127, 255, type=cv2.THRESH_TOZERO)
ret5, zero_inv = cv2.threshold(img, 127, 255, type=cv2.THRESH_TOZERO_INV)
ret6, mask = cv2.threshold(img, 127, 255, type=cv2.THRESH_MASK)
ret7, otsu = cv2.threshold(img, 127, 255, type=cv2.THRESH_OTSU)
ret8, triangle = cv2.threshold(img, 127, 255, type=cv2.THRESH_TRIANGLE)
```

```python
import matplotlib.pyplot as plt

titles = ['Raw', 'Trunc', 'Binary', 'Binary_inv', 'Zero', 'Zero_inv', 'Mask', 'Otsu', 'Triangle']
imgs = [img, trunc, binary, binary_inv, zero, zero_inv, mask, otsu, triangle]

for i in range(9):
    plt.subplot(3, 3, i+1)
    plt.imshow(imgs[i], 'gray')
    plt.title(titles[i])
    plt.xticks([]), plt.yticks([])
plt.show()
```
